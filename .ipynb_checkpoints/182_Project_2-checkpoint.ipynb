```python
import numpy as np
import random
from copy import deepcopy
from tqdm import tqdm
import matplotlib.pyplot as plt
from collections import Counter
import pandas
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
len = 4 #width and height of board
num_spaces = len**2
b = np.zeros((len, len))
```

```python
#board is 2048 board, actions: {1:swipe left, 2:swipe up, 3:swipe right, 4:swipe down}
def step(board, action):
  reward = 0

  #left swipe
  if action == 1:

    #iterate through each row
    for curr in np.arange(0, num_spaces, len):

      #check if neighbor has the same value and combine if they do
      iter = deepcopy(curr)
      for i in np.arange(0,len-1):
        if (board[iter//len][int(iter%len)] == board[(iter+1)//len][int((iter+1)%len)]):
          board[iter//len][int(iter%len)] += board[iter//len][int(iter%len)]
          board[(iter+1)//len][int((iter+1)%len)] = 0
          reward += board[iter//len][int(iter%len)]
        iter += 1

      #shuffle over zeroes from combining
      skips = 0
      iter2 = deepcopy(curr)
      for i in np.arange(0,len):
        if (board[iter2//len][int(iter2%len)] == 0):
          skips += 1
        else:
          temp = board[iter2//len][int(iter2%len)]
          if board[(iter2-skips)//len][int((iter2-skips)%len)] != board[iter2//len][int(iter2%len)]:
            board[iter2//len][int(iter2%len)] = 0
          board[(iter2-skips)//len][int((iter2-skips)%len)] = temp
        iter2 += 1

  #up swipe
  elif action == 2:

    #iterate through each row
    for curr in range(len):

      #check if neighbor has the same value and combine if they do
      iter = deepcopy(curr)
      for i in np.arange(0,len-1):
        if (board[iter//len][int(iter%len)] == board[(iter+len)//len][int((iter+len)%len)]):
          board[iter//len][int(iter%len)] += board[iter//len][int(iter%len)]
          board[(iter+len)//len][int((iter+len)%len)] = 0
          reward += board[iter//len][int(iter%len)]
        iter += len

      #shuffle over zeroes from combining
      skips = 0
      iter2 = deepcopy(curr)
      for i in range(len):#np.arange(curr, (len*(len-1))+curr, len):
        if (board[iter2//len][int(iter2%len)] == 0):
          skips += len
        else:
          temp = board[iter2//len][int(iter2%len)]
          if board[(iter2-skips)//len][int((iter2-skips)%len)] != board[iter2//len][int(iter2%len)]:
            board[iter2//len][int(iter2%len)] = 0
          board[(iter2-skips)//len][int((iter2-skips)%len)] = temp
        iter2 += len
      
  #right swipe
  elif action == 3:

    #iterate through each row
    for curr in np.arange(len-1, len**2, len):

      #check if neighbor has the same value and combine if they do
      iter = deepcopy(curr)
      for i in np.arange(0,len-1):
        if (board[iter//len][int(iter%len)] == board[(iter-1)//len][int((iter-1)%len)]):
          board[iter//len][int(iter%len)] += board[iter//len][int(iter%len)]
          board[(iter-1)//len][int((iter-1)%len)] = 0
          reward += board[iter//len][int(iter%len)]
        iter -= 1

      #shuffle over zeroes from combining
      skips = 0
      iter2 = deepcopy(curr)
      for i in np.arange(0,len):
        if (board[iter2//len][int(iter2%len)] == 0):
          skips += 1
        else:
          temp = board[iter2//len][int(iter2%len)]
          if board[(iter2+skips)//len][int((iter2+skips)%len)] != board[iter2//len][int(iter2%len)]:
            board[iter2//len][int(iter2%len)] = 0
          board[(iter2+skips)//len][int((iter2+skips)%len)] = temp
        iter2 -= 1

  #down swipe
  elif action == 4:

    #iterate through each row
    for curr in np.arange((len**2)-1, len**2-len-1, -1):

      #check if neighbor has the same value and combine if they do
      iter = deepcopy(curr)
      for i in np.arange(0,len-1):
        if (board[iter//len][int(iter%len)] == board[(iter-len)//len][int((iter-len)%len)]):
          board[iter//len][int(iter%len)] += board[iter//len][int(iter%len)]
          board[(iter-len)//len][int((iter-len)%len)] = 0
          reward += board[iter//len][int(iter%len)]
        iter -= len

      #shuffle over zeroes from combining
      skips = 0
      iter2 = deepcopy(curr)
      for i in np.arange(curr+len-1, (len*(len-1))+curr+len, len):
        if (board[iter2//len][int(iter2%len)] == 0):
          skips -= len
        else:
          temp = board[iter2//len][int(iter2%len)]
          if board[(iter2-skips)//len][int((iter2-skips)%len)] != board[iter2//len][int(iter2%len)]:
            board[iter2//len][int(iter2%len)] = 0
          board[(iter2-skips)//len][int((iter2-skips)%len)] = temp
        iter2 -= len

  #assign 2 random spaces to values of 2 and 4 if they are currently zero
  space_1 = random.randint(0, len**2-1)
  space_2 = random.randint(0, len**2-1)
  if board[space_1//len][int(space_1%len)] == 0:
    board[space_1//len][int(space_1%len)] = np.random.choice([2,4], p=[(2/3), (1/3)])
  if board[space_2//len][int(space_2%len)] == 0:
    board[space_2//len][int(space_2%len)] = np.random.choice([2,4], p=[(2/3), (1/3)])

  return reward, board
```

```python
def is_terminal(state):
  indices = np.arange(0,16)
  if 0 in state:
    return False
  for i in range(len**2):
    curr = state[i//len][int(i%len)]
    if (i-1 in indices) and i not in [0,4,8,12]:
      if state[i//len][int((i-1)%len)] == curr:
        return False
    if (i-4 in indices) and i not in [0,1,2,3]:
      if state[(i-4)//len][int(i%len)] == curr:
        return False
    if (i+1 in indices) and i not in [3,7,11,15]:
      if state[i//len][int((i+1)%len)] == curr:
        return False
    if (i+4 in indices) and i not in [12,13,14,15]:
      if state[(i+4)//len][int(i%len)] == curr:
        return False
  
  return True
```

## RL Algorithms Start Here

I implement both semigradient sarsa and N-Step semigradient sarsa.

```python
import torch
import torch.nn as nn

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# Hyper-parameters 
input_size = (len**2)*12
hidden_size = 150
num_classes = 4
alpha = 0.01


class NeuralNet(nn.Module):
    def __init__(self, input_size, hidden_size, num_classes):
        super(NeuralNet, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size) 
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, hidden_size) 
        self.fc3 = nn.Linear(hidden_size, num_classes)

    
    def forward(self, x):
        first = self.fc1(x)
        second = self.relu(first)
        third = self.fc2(second)
        fourth = self.relu(third)
        out = self.fc3(fourth)
        return out


model = NeuralNet(input_size, hidden_size, num_classes).to(device)
criterion = nn.L1Loss()
optimizer = torch.optim.SGD(model.parameters(), lr=alpha) 
```

```python
#convert board state to NN input
def convert(board):
  one_hot_input = np.zeros((12, len**2))
  for i in range(len**2):
    val = int(board[i//len][int(i%len)])
    if val == 0:
      one_hot_input[val][i] = 1
    else:
      one_hot_input[int(np.log2(val))][i] = 1
  return torch.from_numpy(one_hot_input.flatten()).float()

def convert_back(one_hot):
  board = np.zeros((len,len))
  for i in range(len**2):
    arr = one_hot[:,i]
    if np.where(arr == 1)[0] == 0:
      continue
    else:
      board[i//len][int(i%len)] = 2**np.where(arr == 1)[0]
  return board
```

```python
def epsilon_greedy(state, epsilon):
  actions = [model(convert(state).to(device)).detach().numpy()]
  greedy = np.argmax(actions)+1
  if np.random.rand() < 1 - epsilon:
    action = greedy
  else:
    action = random.randint(1,4)
  return action
```

```python

epsilon = 0.01
b = np.zeros((len,len))
max_nums_sarsa = []

for i in tqdm(range(1001)): #update later
  S = deepcopy(b)
  A = random.randint(1,4)
  terminal = False
  R = 0
  n = 0
  while terminal != True:
    R, next_state = step(S, A)
    n += 1
    if is_terminal(S):
      input = convert(S).flatten()
      q = model(input)
      g_approx = torch.FloatTensor([R]).to(device)

      loss = criterion(q, g_approx)
      loss.backward()
      optimizer.step()
      # print("here")
      break

    next_action = epsilon_greedy(next_state, epsilon)
    input_state = convert(S)
    input_next_state = convert(next_state)

    q = model(input_state).to(device)
    g_approx = torch.FloatTensor([R + np.argmax(model(input_next_state).detach().numpy())]).to(device)

    loss = criterion(q, g_approx)
    loss.backward()
    optimizer.step()

    S = next_state
    A = next_action
    
  max_nums_sarsa.append(np.max(S))
```

```python
import matplotlib.pyplot as plt
  
plt.plot(np.arange(0,1001), max_nums_sarsa)
plt.title('SARSA: Max Num Per Iteration')
plt.xlabel('Iteration')
plt.ylabel('Max Num')
plt.show()

count = Counter(max_nums_sarsa)
df = pandas.DataFrame.from_dict(count, orient='index')
df.plot(kind='bar', title = 'SARSA: Max Num Distribution',)
```

```python
model = NeuralNet(input_size, hidden_size, num_classes).to(device)
criterion = nn.L1Loss()
alpha=0.4
optimizer = torch.optim.SGD(model.parameters(), lr=alpha) 
```

```python

epsilon = 0.01
b = np.zeros((len,len))
max_nums_n_sarsa = []
n = 4

for i in tqdm(range(1001)):
    S = deepcopy(b)
    A = epsilon_greedy(S, epsilon)
    T = float("inf")
    t = 0
    
    r_list = []
    s_list = []
    a_list = []
    
    s_list.append(S)
    a_list.append(A)
    while True:
        if t < T:
            R, next_state = step(S, a_list[t])
            r_list.append(R)
            s_list.append(next_state)
            
            if is_terminal(next_state):
                T = t+1
            else:
                next_action = epsilon_greedy(next_state, epsilon)
                a_list.append(next_action)
        tau = t - n + 1
        G = 0
        if tau >= 0:
            #update G
            for i in np.arange(tau+1, min([tau+n, T])):
                G+=r_list[i]
            if (tau+n) < T:
                G += torch.FloatTensor([G + np.argmax(model(convert(s_list[tau+n])).detach().numpy())]).to(device)
                    
                input = convert(s_list[tau]).flatten()
                q = model(input)
                loss = criterion(q, G)
                loss.backward()
                optimizer.step()
        if tau == T-1:
            break
        t += 1
    max_nums_n_sarsa.append(np.max(s_list[-1]))
            
        
        
    
```

```python
plt.plot(np.arange(0,1001), max_nums_n_sarsa)
plt.title('N-Step SARSA: Max Num Per Iteration')
plt.xlabel('Iteration')
plt.ylabel('Max Num')
plt.show()

count = Counter(max_nums_n_sarsa)
df = pandas.DataFrame.from_dict(count, orient='index')
plt = df.plot(kind='bar', title = 'N-Step SARSA: Max Num Distribution')
```
